```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

# EDA

```python
bank.info()
```

```python
# One-Hot Encoding of categrical variables
fin_bank=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome'])
fin_bank
```

```python
# To display all columns(shows all the encoded columns also)
pd.set_option("display.max.columns", None)
fin_bank
```

```python
# Custom Binary Encoding of Binary o/p variables 
fin_bank['default'] = np.where(fin_bank['default'].str.contains("yes"), 1, 0)
fin_bank['housing'] = np.where(fin_bank['housing'].str.contains("yes"), 1, 0)
fin_bank['loan'] = np.where(fin_bank['loan'].str.contains("yes"), 1, 0)
fin_bank['y'] = np.where(fin_bank['y'].str.contains("yes"), 1, 0)
fin_bank
```

```python
# Find and Replace Encoding for month categorical varaible
fin_bank['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
fin_bank=fin_bank.replace(order)
fin_bank
```

```python
fin_bank.info()
```

# Model Building

```python
fin_bank
```

```python
fin_bank.iloc[:,0:11]
```

```python
fin_bank.iloc[:,12:]
```

```python
# Dividing our data into input and output variables
x=pd.concat([fin_bank.iloc[:,0:11],fin_bank.iloc[:,12:]],axis=1) # (All rows, 1-12 columns)+(All rows, from 13 to all columns)
y=fin_bank.iloc[:,11] #12th columns contains the output variable i.e, Y
```

```python
# Logistic regression model
classifier=LogisticRegression()
classifier.fit(x,y)
```

# Model Prediction

```python
# Predict for x dataset
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

# Testing Model Accuracy

```python
# Confusion Matrix for the model accuracy
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39107+1282)/(39107+815+4007+1282)
```

```python
# As accuracy = 0.8933, which is greater than 0.5; Thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0 
classifier.predict_proba(x)[:,1]
```

```python
# ROC Curve plotting and finding AUC value
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
```
